```python
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\f.pranata\Documents\ipho2022.csv')
df
```

```python
# Set index based on rank
df.set_index('Rank', inplace = True)
```

```python
df
```

```python
df.reset_index()
```

```python
# List of top 10 contestants based on rank
df_top_rank = df.sort_values(by = 'Rank', ascending = True)
df_top_rank.head(10)
```

```python
# List of number of contestants based on country
df_contestants = df.groupby('Country').size().sort_values(ascending = False)
df_contestants
```

```python
# Replace NaN countries
df.fillna('Unknown', inplace=True)
```

```python
df
```

```python
# List of award winners from Germany

df_germany = df[df['Country'] == 'Germany']
df_germany

germany_award_counts = df_germany['Award'].value_counts()

print(germany_award_counts)
```

```python
# List of award winners from Malaysia

df_malaysia = df[df['Country'] == 'Malaysia']
df_malaysia

malaysia_award_counts = df_malaysia['Award'].value_counts()
print(malaysia_award_counts)
```

```python
# List of award winners from Indonesia 

df_indonesia = df[df['Country'] == 'Indonesia']
df_indonesia

indonesia_award_counts = df_indonesia['Award'].value_counts()
print(indonesia_award_counts)
```

```python
# Average score of theoretical
df['Theoretical'].mean()
```

```python
# Average score of experimental
df['Experimental'].mean()
```

```python
# Average total score
df['Total'].mean()
```

```python
# Top 10 Contestant with the highest theoretical scores
df_theoretical = df.sort_values(by = 'Theoretical', ascending = False)
df_theoretical.head(10)
```

```python
# Top 10 Contestant with the lowest theoretical scores
df_theoretical_lowest = df.sort_values(by = 'Theoretical', ascending = True)
df_theoretical_lowest.tail(10)
```

```python
# Top 10 contestant with the highest experimental score
df_experimental = df.sort_values(by = 'Experimental', ascending = False)
df_experimental.head(10)
```

```python
# Top 10 contestant with the lowest experimental score
df_experimental_low = df.sort_values(by = 'Experimental', ascending = True)
df_experimental_low.tail(10)
```

```python
# Top 10 contestant with the highest total score

df_highest_total = df.sort_values(by = 'Total', ascending = False)
df_ten_total = df_highest_total.head(10)
```
