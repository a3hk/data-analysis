```python
import pandas as pd

import numpy as np

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\f.pranata\Downloads\Ice Cream Ratings.csv")

df = df.set_index('Date')

df
```

```python
print(plt.style.available)

plt.style.use('seaborn-v0_8-white')
```

# Line plot

```python
df.plot() # line plot by default
```

```python
df.plot(kind = 'line', subplots = True)
```

```python
df.plot(title = 'Ice cream ratings', xlabel = 'Daily rating', ylabel = 'Scores')
```

# Bar plot

```python
df.plot(kind = 'bar', title = 'Ice cream ratings', xlabel = 'Daily rating', ylabel = 'Scores')
```

```python
df['Flavor Rating'].plot(kind = 'bar', title = 'Flavor ratings', xlabel = 'Daily rating', ylabel = 'Scores')
```

# Bar plot stacked

```python
df.plot(kind = 'bar', stacked = True, title = 'Ice cream ratings', xlabel = 'Daily rating', ylabel = 'Scores')
```

```python
df.plot.barh(stacked = True, title = 'Flavor ratings', xlabel = 'Daily rating', ylabel = 'Scores') # bar plot is horizontal
```

# Scatter plot

```python
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 100, c = 'black')
```

# Histogram

```python
df.plot.hist(bins = 20)
```

# Boxplot

```python
df.boxplot()
```

# Area plot

```python
df.plot.area(figsize = (10, 5))
```

# Pie chart

```python
df.plot.pie(y = 'Flavor Rating', figsize = (10, 6), autopct="%1.1f%%")
```
